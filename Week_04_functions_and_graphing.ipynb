Function and a table of (x,y) values

```python
x = 5

y = 4*x + 3 # The function

print(x, ",", y)

print("x \t y")

for x in range(11):
  y = 4*x + 3 # The function
  print(x, "\t", y)
```

Defining a Python function

```python
def f(x):
  y = 4*x +3
  return y

print(5, ",", f(5))

# Loop for al values
for x in range(11):
  print(x, "\t", f(x))
```

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin,xmax], [0,0], 'b') # Blue x axis
plt.plot([0,0], [ymin,ymax], 'b') # blue y axis

plt.show()
```

Graph and table of (x,y) values

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

print("x \t y")
for x in range(xmin, xmax+1):
    y = 0.5*x + 1
    plt.plot([x],[y], 'ro')
    print(x,"\t",y)

plt.show()
```

More ways to customize a graph

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10

points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin,xmax], [0,0], 'b') # x axis
plt.plot([0,0],[ymin,ymax], 'b') # y axis

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Powerful graph")
# ax.grid(True)

ax.set_xticks(np.arange(xmin, xmax, 1))
ax.set_yticks(np.arange(ymin, ymax, 2))

y = 2*x +1
plt.plot(x,y, label='y=2x+1')
plt.plot([4], [6], 'ro', label='point')
plt.plot(x,3*x, label='steeper line')
plt.legend()
plt.show()
```
